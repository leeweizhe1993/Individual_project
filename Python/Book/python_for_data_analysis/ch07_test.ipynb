```python
data = pd.DataFrame(np.arange(6). reshape((2, 3)),
                    index = pd.Index([' Ohio', 'Colorado'], name='state'),
                    columns = pd.Index(['one', 'two', 'three'], name='number'))
print(data, '\n')
print(data.stack(), type(data.stack()), '\n') #stack作用为将列转化为行
print(data.unstack(),type(data.unstack()), '\n')
print(data.stack().unstack()) #两次操作变回原值
```

```python
#按照指定方式将指定数据分组
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
bins = [18, 25, 35, 60, 100]
cats = pd.cut(ages, bins)
cats
```

```python
mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('D://datasets//python_for_data_analysis//datasets//movielens//movies.dat', sep = '::', header = None,
                      names = mnames)
movies[: 10]
```

```python
genre_iter = (set(x.split('|')) for x in movies.genres)
print(genre_iter)
genres = sorted(set.union(*genre_iter))
genres
```

```python
dummies = pd.DataFrame(np.zeros((len(movies), len(genres))), columns = genres)
for i, gen in enumerate(movies.genres):
    dummies.ix[i, gen.split('|')] = 1

movies_windic = movies.join(dummies.add_prefix('Genre_'))
movies_windic.ix[0]
```
