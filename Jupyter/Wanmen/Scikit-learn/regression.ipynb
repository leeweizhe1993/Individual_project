```python
%matplotlib inline
#import numpy as np
#import matplotlib.pyplot as plt
#using linear regression model as estimator
#import sklearn
#from sklearn.linear_model import LinearRegression
#clf = LinearRegression()

#boston house price 
#from sklearn.datasets import load_boston
#data=load_boston()
#n_samples, n_features=data.data.shape
#print n_samples,n_features
#print data.target.shape
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.linear_model import LinearRegression

clf=LinearRegression()
from sklearn.datasets import load_boston
data=load_boston()

n_samples, n_features = data.data.shape

print n_samples, n_features
print data.target.shape
```

```python
#play with the feature
column_i = 5
plt.scatter(data.data[:,column_i], data.target)
print data.feature_names[5]
```

```python
#clf.fit(data.data, data.target)
#predicted = clf.predict(data.data)
#mean_absolute_error(data.target, predicted)
from sklearn.metrics import mean_absolute_error
clf.fit(data.data, data.target)
predicted=clf.predict(data.data)
mean_absolute_error(data.target, predicted)
```

```python
#plot the true price and predicted price
plt.scatter(data.target, predicted)
plt.xlabel('true price')
plt.ylabel('predicted price')
plt.plot(data.target, data.target, color='red')
```

```python
#try another non-linear model
from sklearn.tree import DecisionTreeRegressor

clf2 = DecisionTreeRegressor()
clf2.fit(data.data, data.target)
predicted2=clf2.predict(data.data)
plt.scatter(data.target, predicted2)
plt.xlabel('true price')
plt.ylabel('predicted price')
plt.plot(data.target, predicted2, color='red')
```

```python
#overfit
from sklearn.metrics import mean_absolute_error
mean_absolute_error(data.target, predicted2)
```
