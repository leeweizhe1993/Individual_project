```python
import random
import pandas as pd
from IPython.display import display
from sklearn.utils import shuffle
```

```python
hu = pd.read_csv("Hu.csv", )
lu = pd.read_csv("Lu.csv")
hu.drop(['序号'], axis = 1, inplace = True), lu.drop(['序号'], axis = 1, inplace = True)
hu['导师'] = '胡'
lu['导师'] = '陆'
display(hu.head(), lu.head())
```

```python
total = shuffle(pd.concat([hu, lu], ignore_index=True))
```

```python
total["性别"].value_counts()
```

```python
total["导师"][:5].value_counts()['陆']
```

```python
def shuffle_list(names):
    first = [x for x in range(0, 70, 5)]
    state = 0
    while state == 0:
        names = shuffle(names)
        names.index = range(len(names))
        for i in first:
            if len(list(names["性别"][i:i + 5].value_counts())) < 2:
                state += 1
            elif len(list(names["导师"][i:i + 5].value_counts())) < 2 or names["导师"][i:i + 5].value_counts()['陆'] > 2:
                state += 1
        if state == 0:
            return names
        else:
            state = 0
        

result = shuffle_list(total)
```

```python
a = [x for x in range(1, 15)]
result.index = list(map(str, sum(list(map(list,zip(a, a, a, a, a))), [])[:-1]))
display(result[:25])
display(result[25:50])
display(result[50:])
```
