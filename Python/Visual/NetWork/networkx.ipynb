# Node
节点（顶点）的集合以及标识的节点对（称为边，链接等）。在NetworkX中，节点可以是任何可清除对象，例如文本字符串，图像，XML对象，另一个图形，自定义节点对象等

```python
G = nx.Graph()    #新建空图
G.add_node('Node-1')    #添加1个节点
```

```python
G.add_nodes_from(['Node-2', 'Node-3'])    #添加多个节点
```

```python
H = nx.path_graph(10)    #type(H) networkx.classes.graph.Graph
G.add_nodes_from(H)    #将H中的许多结点作为G的节点
# G.add_node(H)    #将H作为G中的一个节点
```

```python
G.node()
```

```python
H.node()
```

```python
G[1]    #查看连接
```

```python
for n, nbs in G.adj.items():
    print(nbs)
```

# Edges

`G`能够一次增加一条边

```python
G.add_edge(1, 2)    #只能增加边，有属性，除非指定属性名和值“属性名=值”
e = (2, 3)
G.add_edge(*e)   #注意！ G.add_edge(e)会报错！
```

```python
G.add_edges_from([(1, 2), (1, 3)])    #一次添加多个边
```

增加 ebunch边。ebunch：包含边元组的容器，比如序列、迭代器、文件等。  
这个元组可以是2维元组或 三维元组 （node1，node2，an_edge_attribute_dictionary）,an_edge_attribute_dictionary比如{‘weight’:3.1415}

```python
G.add_edges_from(H.edges())    #从其他图中添加边
```

```python
G.number_of_nodes(), H.number_of_nodes()
```

```python
G.edges()
```

# Visualize

```python
G.edges[1, 3]['color'] = 'green'
G.edges[3, 4]['color'] = 'red'
```

```python
pos = nx.spring_layout(G)
nx.draw(G, cmap = plt.get_cmap('rainbow'), with_labels=True, pos=pos)
```

# 有向图

```python
dg = nx.DiGraph()
dg.add_weighted_edges_from([(1, 2, 0.5), (3, 1, 0.9)])
dg.out_degree(1, weight='weight')
```

```python
dg.nodes()
```

```python
list(dg.successors(1)),list(dg.predecessors(1))     #获取子节点/父节点
```

```python
list(dg.successors(1)) == list(dg.neighbors(1))
```

```python
nx.draw(dg, with_labels=True)
```
