```python
class Node(object):    #定义node
    
    def __init__(self, value = None, pre = None, nxt = None):    #值和指针
        self.value, self.pre, self.next = value, pre, nxt
        
class CircularDoubleLinkedList(object):
    """循环双端链表 ADT
    多了个循环其实就是把 root 的 pre 指向 tail 节点，串起来
    """
    
    def __init__(self, maxsize = None):
        self.maxsize = maxsize
        node = Node()
        node.next, node.pre = node, node    #指向自己实现闭环
        self.root = node
        self.length = 0
        
    def __len__(self):
        return self.length
    
    def headnode(self):
        return self.root.next
    
    def tailnode(self):
        return self.root.pre    #循环列表，root节点的pre指向尾节点
    
    def append(self, value):
        if self.maxsize is not None and len(self) > self.maxsize:
            raise Exception("full")
        node = Node(value = value)    #新建一个节点
        tailnode = self.tailnode()
        
        tailnode.next = node    #append的节点增添至尾节点后
        node.pre = tailnode    #之前的tialnode作为pre节点
        node.next = self.root    #append节点后为root节点
        self.root.pre = node    #更新root节点的pre
        
        self.length += 1
        
    def appendleft(self, value):
        if self.maxsize is not None and len(self) > self.maxsize:
            raise Exception("full")
        node = Node(value = value)
        
        if self.root.next is self.root:    #空链表
            node.pre = self.root
            node.next = self.root
            self.root.pre = node
            selg.root.next = node
        else:
            node.pre = self.root
            headnode = self.root.next
            node.next = headnode
            headnode.pre = node
            self.root.next = node
        self.length += 1
        
    def remove(self, node):    # O(1)，传入node 而不是 value 就能实现 O(1) 删除
        """remove
        :param node  # 在 lru_cache 里实际上根据key 保存了整个node:
        """
        if node is self.root:
            return
        else:
            node.pre.next = node.next #将前一节点与后一节点相连
            node.next.pre = node.pre
        self.length -= 1
        return node
    
    def iter_node(self):
        if self.root.next is self.root:
            return
        curnode = self.root.next
        while curnode.next is not self.root:
            yield curnode
            curnode = curnode.next
        yield curnode
        
    def __iter__(self):
        for node in self.iter_node():
            yield node.value
            
    def reverse_iter(self):
        '''相比单链表独有的反序遍历'''
        if self.root.pre is self.root:
            return
        curnode = self.root.pre
        while curnode.pre is not self.root:
            yield curnode
            curnode = curnode.pre
        yield curnode
        
class Deque(CircularDoubleLinkedList):
    
    def pop(self):
        if len(self) == 0:
            raise Exception("Empty")
        tailnode = self.tailnode()
        value = tailnode.value
        self.remove(tailnode)
        return value
    
    def popleft(self):
        if len(self) == 0:
            raise Exception("Empty")
        headnode = self.headnode()
        value = headnode.value
        self.remove(headnode)
        return value
    
class Stack(object):
    
    def __init__(self):
        self.deque = Deque()
        
    def push(self, value):
        return self.deque.append(value)
    
    def pop(self):
        return self.deque.pop()
    
    def __len__(self):
        return len(self.deque)
    
    def is_empty(self):
        return len(self.deque) == 0
    
def test_stack():
    s = Stack()
    for i in range(5):
        s.push(i)
        
    assert len(s) == 5
    assert s.pop() == 4
    assert s.pop() == 3
    assert len(s) == 3
    
    for i in range(3):
        s.pop()
        
    assert s.is_empty()
    
    import pytest
    with pytest.raises(Exception) as excinfo:
        s.pop()
    assert "Empty" in str(excinfo.value)
    print(excinfo.value)
    
if __name__ == "__main__":
    test_stack()
```
