# 提取行为数据

```python
import datetime, time, re, os
import numpy as np
import pandas as pd
from IPython.core.debugger import set_trace
```

```python
def time_count(record):
    '''时长统计'''
    pattern_start = re.compile(': \(([\s\S]*?)\)')
    pattern_end = re.compile('Time="([\s\S]*?)"')
    if '    在“无标题 - Google Chrome”中用户键盘输入 [... Enter]\n' in record:
        start_time = re.findall(pattern_start,
                                record[record.index('    在“无标题 - Google Chrome”中用户键盘输入 [... Enter]\n') - 1])[0]
    elif '    在“无标题 - Google Chrome”中用户键盘输入 [... Enter ... Ctrl-V Enter Enter]\n' in record:
        start_time = re.findall(pattern_start,
                                record[record.index('    在“无标题 - Google Chrome”中用户键盘输入 [... Enter ... Ctrl-V Enter Enter]\n') - 1])[0]
    elif '    在“无标题 - Google Chrome”中用户键盘输入 [... Enter ... Ctrl-V]\n' in record:
        start_time = re.findall(pattern_start,
                                record[record.index('    在“无标题 - Google Chrome”中用户键盘输入 [... Enter ... Ctrl-V]\n') - 1])[0]
    elif '    在“无标题 - Google Chrome”中用户键盘输入 [... Enter ... Ctrl-V Enter]\n' in record:
        start_time = re.findall(pattern_start,
                                record[record.index('    在“无标题 - Google Chrome”中用户键盘输入 [... Enter ... Ctrl-V Enter]\n') - 1])[0]
    elif '    在“New Tabs - Google Chrome”中用户键盘输入 [... Enter]\n' in record:
        start_time = re.findall(pattern_start,
                                record[record.index('    在“New Tabs - Google Chrome”中用户键盘输入 [... Enter]\n') - 1])[0]
    else:
        start_time = re.findall(pattern_start,
                                record[record.index('    在“New Tabs - Google Chrome”中用户键盘输入 [... Enter ... Ctrl-V Enter Enter]\n') - 1])[0]
    start_time = datetime.datetime.strptime(start_time,'%Y/%m/%d %H:%M:%S')
    if '        <Text>Esc</Text>\n' in record:
        end_time = re.findall(pattern_end, record[record.index('        <Text>Esc</Text>\n') - 9])[0]
        end_time = datetime.datetime.strptime(end_time,'%H:%M:%S')
    elif '        <Text>... Enter Esc</Text>\n' in record:
        end_time = re.findall(pattern_end, record[record.index('        <Text>... Enter Esc</Text>\n') - 9])[0]
        end_time = datetime.datetime.strptime(end_time,'%H:%M:%S')
    elif '        <Text>... Esc Esc</Text>\n' in record:
        end_time = re.findall(pattern_end, record[record.index('        <Text>... Esc Esc</Text>\n') - 9])[0]
        end_time = datetime.datetime.strptime(end_time,'%H:%M:%S')
    else:
        end_time = re.findall(pattern_end, record[record.index('        <Text>... Esc</Text>\n') - 9])[0]
        end_time = datetime.datetime.strptime(end_time,'%H:%M:%S')
    return (end_time - start_time).seconds

def task_count(record):
    '''步骤统计'''
    start = record.index(
    '操作系统: 7601.24387.amd64fre.win7sp1_ldr_escrow.190305-1700 6.1.1.0.2.48\n')
    stop = start + record[start:].index('    </textarea>\n')
    # set_trace()
    steps = ''.join(record[start:stop])
    start_task = steps[steps.index('[... Enter') - 100: steps.index('[... Enter')]
    end_task = steps[steps.index('Esc]\n') - 100: steps.index('Esc]\n')]

    pattern_step = re.compile('问题步骤 ([\s\S]*?):')
    start_step = re.findall(pattern_step, start_task)[0]
    esc_step = re.findall(pattern_step, end_task)[0]

    return int(esc_step) - int(start_step)
```

```python
all_trial = pd.DataFrame(columns = ['sub', 'trial', 'web', 'time', 'control'])

for i in os.listdir('Screen-record'):
    if 'mht' in i:
        with open('Screen-record//1//' + i, 'r') as f:
            record = list(f)
            f.close()
        new = pd.DataFrame(columns = ['sub', 'trial', 'web', 'time', 'control'])
        new.loc[0] = [i[1:-6], i[-6], i[-5], time_count(record), task_count(record)]
        all_trial = all_trial.append(new, ignore_index=True)
all_trial.to_csv("behavior1.csv", index = False)
all_trial.head()
```

# 提取眼动数据

```python
message = pd.read_csv('eye-data//1//Message.csv')
behavior = pd.read_csv('behavior1.csv')
trial = pd.read_csv('eye-data//1//Trial.csv')
message.head(10)
```

```python
behavior = pd.concat([behavior, pd.DataFrame(columns=['blink', 'fix', 'med_dur', 'med_sac'])],
                     sort = False)

for i in range(len(message)):
    sub = int(message.loc[i][0][:2])
    if 'jd' in message.loc[i][-1]:
        site = 'j'
    else:
        site = 't'
    if 'Control' in message.loc[i][-1]:
        cond = 'c'
    else:
        cond = 'r'
    blink = trial['BLINK_COUNT'][i]
    fix = trial['FIXATION_COUNT'][i]
    med_dur = trial['MEDIAN_FIXATION_DURATION'][i]
    med_sac = trial['MEDIAN_SACCADE_AMPLITUDE'][i]
    num = behavior[(behavior['sub'] == sub) & (behavior.trial == cond) & (behavior.web == site)].index[0]
    behavior.loc[[num],['blink', 'fix', 'med_dur', 'med_sac']] = [blink, fix, med_dur, med_sac]
behavior.to_csv('behavior1.csv', index = False)
behavior.head()
```
