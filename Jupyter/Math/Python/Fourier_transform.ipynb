```python
g1=np.array([1,1,1,1,1,1,1,1]) #时域值
g2=np.array([1,2,1,2,1,2,1,2]) #时域值
fftg1=np.fft.rfft(g1)   #频域值
fftg2=np.fft.rfft(g2)   #频域值
print(fftg1)
print(fftg2)

plt.figure()
plt.subplot(121)
plt.plot(abs(fftg1)) #一直没搞懂求模值为什么是abs而不是norm
plt.subplot(122)
plt.plot(abs(fftg2))
plt.show()
```
